Gradient Boost Model Algorithm:
The model is initialized with default hyperparameters and trained using the training dataset. fit ensures the model learns the patterns from the data.

```python
import numpy as np

class DecisionTree:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth
        self.feature_index = None
        self.threshold = None
        self.left_subtree = None
        self.right_subtree = None
        self.value = None

    def fit(self, X, y, depth=0):
        unique_classes = np.unique(y)

        if len(unique_classes) == 1 or depth == self.max_depth:
            self.value = np.mean(y)
            return

        num_features = X.shape[1]
        best_feature, best_threshold = None, None
        best_score = float('inf')

        for feature_index in range(num_features):
            thresholds = np.unique(X[:, feature_index])
            for threshold in thresholds:
                left_mask = X[:, feature_index] <= threshold
                right_mask = ~left_mask

                if np.sum(left_mask) > 0 and np.sum(right_mask) > 0:
                    score = self._calculate_score(y[left_mask], y[right_mask])
                    if score < best_score:
                        best_score = score
                        best_feature = feature_index
                        best_threshold = threshold

        if best_score == float('inf'):
            self.value = np.mean(y)
        else:
            self.feature_index = best_feature
            self.threshold = best_threshold
            left_mask = X[:, best_feature] <= best_threshold
            self.left_subtree = DecisionTree(max_depth=self.max_depth)
            self.left_subtree.fit(X[left_mask], y[left_mask], depth + 1)

            right_mask = ~left_mask
            self.right_subtree = DecisionTree(max_depth=self.max_depth)
            self.right_subtree.fit(X[right_mask], y[right_mask], depth + 1)

    def _calculate_score(self, left_y, right_y):
        return len(left_y) * np.var(left_y) + len(right_y) * np.var(right_y)

    def predict(self, X):
        if self.value is not None:
            return np.full(X.shape[0], self.value)
        else:
            left_mask = X[:, self.feature_index] <= self.threshold
            right_mask = ~left_mask
            result = np.zeros(X.shape[0])
            result[left_mask] = self.left_subtree.predict(X[left_mask])
            result[right_mask] = self.right_subtree.predict(X[right_mask])
            return result


class GradientBoosting:
    def __init__(self, n_estimators=100, learning_rate=0.1, max_depth=3):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.trees = []

    def fit(self, X, y):
        # Initialize the model with the mean of the target values
        initial_prediction = np.mean(y)
        residuals = y - initial_prediction

        for _ in range(self.n_estimators):
            tree = DecisionTree(max_depth=self.max_depth)
            tree.fit(X, residuals)
            self.trees.append(tree)
            residuals -= self.learning_rate * tree.predict(X)

    def predict(self, X):
        predictions = np.sum([tree.predict(X) for tree in self.trees], axis=0)
        return np.round(predictions)
```

# **Model**
Trying with different n_estimators

Initialization and Residuals Calculation:

The GradientBoosting class initializes with the number of estimators, learning rate, and maximum tree depth.
The first prediction is the mean of the target values (y), and residuals are calculated by subtracting this initial prediction from y.
Fitting Trees on Residuals:

For each estimator, a new DecisionTree is trained to predict the residuals from the previous iteration.
Each tree is added to the trees list after being fitted.
Updating Residuals:

After each tree is trained, predictions are made, and the residuals are updated by subtracting the product of the learning rate and the tree’s predictions.
This iterative approach ensures each tree focuses on the remaining errors.
Prediction Aggregation:

During prediction, each tree contributes to the final prediction. The outputs of all trees are summed, and the aggregated result represents the final model prediction.
Output Rounding for Classification:

The sum of the tree predictions is rounded to the nearest integer, converting continuous outputs into class labels, suitable for binary classification or regression tasks.
